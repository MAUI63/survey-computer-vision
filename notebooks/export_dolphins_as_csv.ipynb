```python
import json
import pandas as pd

with open("/workspaces/cv/data/ml/annotated/202505-dolphins/all.json") as f:
    dolphins = json.load(f)

rows = []
for img in dolphins["images"]:
    meta = img["meta"]
    location = meta["location"]
    if location is None:
        location = {}
    rows.append(
        dict(
            file_name=img["file_name"],
            survey=meta["survey"],
            capture_time=meta["capture_time"],
            latitude=location.get("latitude", None),
            longitude=location.get("longitude", None),
            altitude=location.get("altitude", None),
        )
    )
df = pd.DataFrame(rows)
df
```

```python
df.to_csv("/workspaces/cv/.tmp/dolphins.csv", index=False)
```
